#### Multiplying a list by a scalar

```
# using for loop
#
L2 = []
scalar = 2

for e in L:
  L2.append(e * scalar)

print(f'original list -> {L}, multiplied by 2 -> {L2}')
```

```
# using list comprehension
#
L2 = [e * scalar for e in L]
print(f'original list -> {L}, multiplied by 2 -> {L2}')
```

### 2.5 Speed Test

```
from datetime import datetime
```

```
a = np.random.randn(100)
b = np.random.randn(100)
T = 100000
```

```
def slow_dot_product(a,b):
  result = 0
  for e,f in zip (a,b):
    result += e * f
  return result
```

```
t0 = datetime.now()
for t in range(T):
  slow_dot_product(a,b)
dt1 = datetime.now() - t0

t0 = datetime.now()
for t in range(T):
  a.dot(b)
dt2 = datetime.now() - t0
print(f'dt1/dt2 = {dt1.total_seconds()/dt2.total_seconds()}')
```

```
from scipy.stats import norm
```

```
plt.plot(x,fx);
```

```
Fx = norm.cdf(x, loc=0, scale=1)
```

```
plt.plot(x,Fx);
```

```
logx = norm.logpdf(x, loc=0, scale=1)
```

```
plt.plot(x,logx);
```
